##wash dataset for visulzation

Manipulation of the Artists born(place and date)set. Here we creat a timeline from 1755-193 and each interval is 15 years. The value in the timeline will be set to 1 if the Artist is alive at that time interval.

```python
artistBirth = pd.read_csv("artistBirthDeath19thLocation.csv")
```

```python
location=artistBirth['location']
```

```python
latitude=[]
longitude=[]
```

```python
for i in range(len(location)):
  location[i]=location[i][1:]
  location[i]=location[i][:-1]
  location[i]= location[i].split(", ")
  latitude.append(location[i][0])
  longitude.append(location[i][1])
```

```python
print(latitude[0],longitude[0])
```

```python
artistBirth['latitude']=latitude
artistBirth['longitude']=longitude
```

```python
print((location[0]))
```

```python
artistBirth.to_csv('artistBirthDeath19thLocation.csv')
```

```python
df = pd.DataFrame()
years = [1755,1770,1785,1800,1815,1830,1845,1860,1875,1890,1905]
df['name']=artistBirth['name']
```

```python
print(df)
```

```python
for i in range(len(artistBirth)):
  birth = artistBirth.loc[i]['birthdate']
  death = artistBirth.loc[i]['deathdate']
  timeline=[]
  for j in range(len(years)):
    if(years[j]<birth):
      if(years[j]+14>=birth):
        timeline.append(2)
      else:
        timeline.append(0)
    else:
      if(years[j]>death):
        timeline.append(0)
      else:
        timeline.append(1)
  for j in range(len(years)):
    df.at[i,years[j]]=timeline[j]
  
```

```python
print(df)
```

```python
df['latitude']=latitude
df['longitude']=longitude
```

```python
df.to_csv('lifeTimeline.csv')
```

##Visulization
With the help of pyecharts, we generated a interactive Map with timeline to visualize the birth location and the life time of Artists. Besides that, we also manually choose some significant History Events and add them in the corresponsible Map, to makke the reference of what is happening at that time.

```python
import pandas as pd
import numpy as np
from numpy import array
import math
```

```python
import heapq
from pyecharts import options as opts
#from pyecharts import Style
from pyecharts.globals import ThemeType
from pyecharts.commons.utils import JsCode
from pyecharts.charts import Timeline, Grid, Bar, Map, Pie, Geo
from pyecharts.globals import GeoType
import csv
```

```python
timeline = Timeline(init_opts=opts.InitOpts(width='1200px', height='600px')) 
timeline.add_schema(axis_type='category', orient='horizontal', symbol='circle', symbol_size=10, 
                    play_interval=4000, is_auto_play=True, is_loop_play=True, is_timeline_show=True, 
                    control_position='left', pos_left='10%', pos_bottom='5%', width='800px', height='40px',
                    linestyle_opts = opts.LineStyleOpts(width = 5,
                                   type_ = 'dashed',
                                   color = '#564537'),
                    label_opts = opts.LabelOpts(color = '#564537',
                                    font_size = 15),
                    itemstyle_opts=opts.ItemStyleOpts(color="#564537")

                    ) 
```

```python
#define dataset
df = pd.read_csv('lifeTimeline.csv')
```

```python
geo_cities_coords={i:[df.iloc[i]['longitude'],df.iloc[i]['latitude']]
                    for i in range(len(df))} 
```

```python
print(df.iloc[0]['name'])
```

```python
# df.iloc[j][str(i)]
```

```python
df2=pd.read_csv('HistoryTimeline_15samples.csv')
```

```python
geo_events_coords={i:[df.iloc[i]['longitude'],df.iloc[i]['latitude']]
                    for i in range(len(df2))} 
```

```python
for i in range(1755, 1920,15):
    g = Geo(init_opts=opts.InitOpts(bg_color="#dfc5ae", width="1000px", height="600px"))
    g.add_schema(maptype="world",
                 itemstyle_opts=opts.ItemStyleOpts(color="#e2c29d", #background color
                                                   color0="#72634a",
                                                  border_color="#564537"))
    #add lattitude/longitude
    for k, v in list(geo_cities_coords.items()):
      if(df.iloc[k][str(i)]>0):
        g.add_coordinate(df.iloc[k]['name'], v[0], v[1])
    for k, v in list(geo_events_coords.items()):
      if(df2.iloc[k][str(i)]>0):
        g.add_coordinate(df2.iloc[k]['name'], v[0], v[1])
    data_pair=[]
    for j in df.index:
      if df.iloc[j][str(i)]>0:
        data_pair.append([df.iloc[j]['name'],df.iloc[j][str(i)]])
    #data_pair = [(df.iloc[j]['name'],df.iloc[j][str(i)]*10) for j in df.index]
    data_pair2=[]
    for j in df2.index:
      if df2.iloc[j][str(i)]>0:
        data_pair2.append([df2.iloc[j]['name'],df2.iloc[j][str(i)]])
    g.add('Artist',data_pair, 
          type_=GeoType.EFFECT_SCATTER, 
          color="#392e25",
          #geo_normal_color="#a2947c",
          symbol_size=5)
    g.add('History Events',data_pair2, 
          type_=GeoType.EFFECT_SCATTER, 
          color="#87CEFA",
          #geo_normal_color="#a2947c",
          symbol_size=20)
    g.set_series_opts(label_opts=opts.LabelOpts(is_show=True, 
                                                position='bottom',
                                                formatter='{b}',
                                                font_size=10,
                                                color='#4c3e31',))
    g.set_global_opts(title_opts=opts.TitleOpts(title="Life Timeline of 19th century Artists",
                                                subtitle="what happened in{start}-{end}".format(start=i, end=i+14),
                                                title_textstyle_opts=opts.TextStyleOpts(color="#022955")))
    
    timeline.add(g, "{}".format(i))
```

```python
timeline.render("timeline_map.html")
```
